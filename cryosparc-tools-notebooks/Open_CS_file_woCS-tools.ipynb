```python
import numpy as np # cs files are structured numpy array
import pandas as pd # used to view inputed star and cs files
```

```python
# edit to point to cs particles file
csfile='/mnt/McQueen-001/CS-PROJECTS-McQueen-001/CS-XXX/J69/P20_S2_live_particles.cs'
```

```python
def cs_to_dataframe(cs_file):
    # load cs file as numpy array (organised in rows)
    a = np.load(csfile)
    # get names of columns
    cols=list(a.dtype.names)
    # load numpy array as dataframe (see https://pbpython.com/pandas-list-dict.html)
    return pd.DataFrame.from_records(a.tolist(), columns=cols)
 
    
```

```python
cs_as_df = cs_to_dataframe(csfile)
pd.set_option('display.max_columns', 100)
display(cs_as_df)
#pd.set_option('display.max_rows', 10000)
#pd.set_option('display.expand_frame_repr', True)
```
